```python
import circle_fit as cf
import matplotlib.pyplot as plt
import numpy as np

def get_unit_circle():
    phasor = np.e**(1j*np.linspace(0, 2*np.pi, 100))
    phasor *= np.random.uniform(0.9, 1.1, 100)
    phasor += 1 + 1j
    x = phasor.real
    y = phasor.imag
    return x, y

x, y = get_unit_circle()
plt.scatter(x, y)
```

```python
coords = np.column_stack((x, y))
```

```python
xc,yc,r,s = cf.hyperLSQ(coords)
```

```python
cf.plot_data_circle(coords,xc,yc,r)
```

```python
xc, yc, r, s = cf.standardLSQ(coords)
```

```python
cf.plot_data_circle(coords,xc,yc,r)
```

```python
xc, yc, r, s = cf.prattSVD(coords)
cf.plot_data_circle(coords, xc, yc, r)
```

```python
xc, yc, r, s = cf.riemannSWFLa(coords)
cf.plot_data_circle(coords, xc, yc, r)
```

```python
xc, yc, r, s = cf.lm(coords, par_ini=np.array([xc, yc, r]))
cf.plot_data_circle(coords, xc, yc, r)
```
